```python
def solucion_triang_inf(L,b):
  filaSize = len(L)
  res = np.zeros((filaSize, 1))

  for i in range(filaSize):
    j = 0
    while (j <= i):
      res[i] = res [i] + L[i][j] * b[i]
      j+=1
  
  return res



# Definir la matriz L y el vector b
L = np.array([[1, 0, 0], [2, 3, 0], [4, 5, 6]])
b = np.array([1, 2, 3]).transpose()
result = solucion_triang_inf(L,b)

print(result) 
```
